```python
with pdfplumber.open(f"{path}{docs[0]}") as pdf:
    page = pdf.pages[0]
    max_size = 0
    title = ''
    lines = page.extract_text_lines(x_tolerance=1)
    for line in lines[:8]:
        if line['chars'][0]['size'] > max_size:
            title = line['text']
            max_size = line['chars'][0]['size']
        elif line['chars'][0]['size'] == max_size:
            title += f" {line['text']}"


    text = page.search('(?s)(?<=Abstract).*?(?=(Keywords|Introduction))',x_tolerance=1)[0]['text']
    print(text)

```

```python
table = []
for doc in docs:
  if doc.endswith('.pdf'):
    with pdfplumber.open(f"{path}{doc}") as pdf:
        try:

            page = pdf.pages[0]
            max_size = 0
            title = ''
            lines = page.extract_text_lines(x_tolerance=1)
            for line in lines[:8]:
                if line['chars'][0]['size'] > max_size:
                    title = line['text']
                    max_size = line['chars'][0]['size']
                elif line['chars'][0]['size'] == max_size:
                    title += f" {line['text']}"


            text = page.search('(?s)(?<=(Abstract|ABSTRACT)).*?(?=(Keywords|Introduction|1. Introduction|1. INTRODUCTION|1.Introduction))',x_tolerance=1)[0]['text']
            table.append({'title':title,'abstract':text,'doc':doc})
        except:
            pass
```

```python
filtered = [p['doc'] for p in table]
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
for doc in table:
    doc['abstract'] = doc['abstract'].lower()
    doc['abstract-minus-stopwords'] = ' '.join([word for word in doc['abstract'].split() if word not in (stopwords.words('english'))])
```

```python
table[4]['abstract-minus-stopwords']
```

```python
import nltk
nltk.download('wordnet')
```

```python
nltk.download('omw-1.4')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
for doc in table:
    doc['lemma-abstract'] = ' '.join([lemmatizer.lemmatize(word) for word in doc['abstract-minus-stopwords'].split()])
```

```python
table[4]['lemma-abstract']
```

```python
from nltk.tokenize import word_tokenize

nltk.download('punkt')
```

```python
import re
```

```python
for doc in table:
    doc['lemma-abstract'] = re.sub(r'([^a-zA-Z ]+?)', '', doc['lemma-abstract'])

```

```python
table[4]['lemma-abstract']
```

```python
data = []

for doc in table:
    data.append(doc['lemma-abstract'])
```

```python
tokenized_data = [word_tokenize(abstract) for abstract in data]
```

```python
from gensim.models import Word2Vec
import gensim.downloader
```

```python
vecs = gensim.downloader.load('conceptnet-numberbatch-17-06-300')
```

```python


model = Word2Vec(
    sentences=tokenized_data,
    vector_size=100,  # You can adjust the size of the word vectors
    window=5,  # Context window size
    min_count=1,  # Minimum frequency of a word to be included
    sg=0,  # CBOW model (0 for CBOW, 1 for Skip-gram)
)
```

```python
model.build_vocab(tokenized_data)
```

```python
model.train(tokenized_data, total_examples=len(tokenized_data), epochs=10)
```

```python
document_embeddings = [sum(model.wv[word] for word in abstract) / len(abstract) for abstract in tokenized_data]
```

```python
len(document_embeddings[1])
```

```python
for i in range(len(table)):
    table[i]['embedding'] = document_embeddings[i]
```

```python
csv = {}
for doc in table:
    csv[doc['doc']] = {i:doc[i] for i in doc if i!='doc'}
```
